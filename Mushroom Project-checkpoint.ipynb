```python
from sklearn.preprocessing import StandardScaler
```

```python
X=ds.drop(['class'], axis=1)
Y=ds['class']
```

```python
scale=StandardScaler()
X=scale.fit_transform(X)
```

```python
X_train, X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.1)
```

```python
clf = DecisionTreeClassifier()
clf = clf.fit(X_train, Y_train)
clf.score(X_train,Y_train)
predclf=clf.predict(X_test)
print(accuracy_score(Y_test,predclf))
print(confusion_matrix(Y_test,predclf))
print(classification_report(Y_test,predclf))
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,Y_train)
lr.score(X_train,Y_train)
predlr=lr.predict(X_test)
print(accuracy_score(Y_test,predlr))
print(confusion_matrix(Y_test,predlr))
print(classification_report(Y_test,predlr))
```

```python
gnb=GaussianNB()
gnb.fit(X_train,Y_train)
predgnb=gnb.predict(X_test)
print(accuracy_score(Y_test,predgnb))
print(confusion_matrix(Y_test,predgnb))
print(classification_report(Y_test,predgnb))
```

```python
svc=SVC(kernel='rbf')

svc.fit(X_train,Y_train)
svc.score(X_train,Y_train)
predsvc=svc.predict(X_test)
print(accuracy_score(Y_test,predsvc))
print(confusion_matrix(Y_test,predsvc))
print(classification_report(Y_test,predsvc))
```

```python
Y_pred=clf.predict(X_test)
```

```python
print("Decision Tree Classifier report \n", classification_report(Y_test, Y_pred))
```

```python
cfm=confusion_matrix(Y_test, Y_pred)

sns.heatmap(cfm, annot = True,  linewidths=.5, cbar =None)
plt.title('Decision Tree Classifier confusion matrix')
plt.ylabel('True label')
plt.xlabel('Predicted label')
```
